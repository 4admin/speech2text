```python
modelo = whisper.load_model('small') #A escala aqui é tiny, base, small, medium, large - Menos processamento vai mais rápido, mas erra mais. Mais processamento emora, mas erra menos
#No modo 'small' abri MP4 de 12GB e uso 5GB e de RAM
resposta = modelo.transcribe('Subirá1.mp3')
faladf = pd.DataFrame.from_dict(resposta['segments'])
faladf.head()
```

```python
faladf['text'].to_csv('arquivo.txt', index=False, header=False)
```
