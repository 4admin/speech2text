```python
modelo = whisper.load_model('large') #A escala aqui é tiny, base, small, medium, large - Menos processamento vai mais rápido, mas erra mais. Mais processamento emora, mas erra menos
#No modo 'small' abri MP4 de 12GB e uso 5GB e de RAM
resposta = modelo.transcribe('/content/gdrive/My Drive/Colab Notebooks/POC/10-08-23 - PROVA DE CONCEITO.mp4')
faladf = pd.DataFrame.from_dict(resposta['segments'])
```
